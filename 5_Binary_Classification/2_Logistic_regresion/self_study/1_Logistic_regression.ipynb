```python

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LogisticRegression
```

```python

N = 1000
data = pd.DataFrame({
    "x1": np.append(np.random.normal(1.5, 0.5, N), np.random.normal(3, 0.5, N)),
    "x2": np.append(np.random.normal(1.5, 0.5, N), np.random.normal(3, 0.5, N)),
    "y": np.append([0] * N, [1] * N)
})
```

```python

fig, ax = plt.subplots(figsize=(8,8))
ax.scatter(data.x1, data.x2, c=data.y, cmap='bwr')
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_xlim(0, 5)
ax.set_ylim(0, 5)
```

```python

clf = LogisticRegression()
clf.fit(data.iloc[:, :-1], data.iloc[:, -1])
t1, t2 = clf.coef_[0]
t0 = clf.intercept_
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.scatter(data.x1, data.x2, c=data.y, cmap='bwr')
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_xlim(0, 5)
ax.set_ylim(0, 5)
ax.plot([0, -t0/t2],[-t0/t1,0], "black")
```
