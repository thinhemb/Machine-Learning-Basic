### A.  Dữ liệu phân loại Categorical data 

1. Dữ liệu phân loại là gì?

Là thuộc tính dữ liệu giúp cho việc phân loại riêng biệt của một tập các danh mục hoặc lớp. Có 2 loại chính trong dữ liệu phân loại đó là: danh nghĩa và thứ tự.

Đối với loại dữ liệu danh nghĩa thì sẽ không có khái niệm thứ tự: chẳng hạn như danh mục thời tiết gồm có: gió, mưa, mây, nắng, lạnh, tuyết. Thì không có loại nào đứng trước loại nào.

Ví dụ: Thể loại phim: Cổ trang, Hàn Quốc, Lãng mạn, Trinh Thám, Cảnh sát hình sự, Tài liệu.....

Còn đối với loại dữ liệu phân loại loại thứ tự. 

Ví dụ: Size của áo: XS, S, M, L, XL, XXL  


2. Xử lí dữ liệu phân loại thế nào?

Chuẩn hóa dữ liệu danh mục chúng ta áp dụng mô hình one hot encoding tương tự như việc mã hóa nhãn của bài toán phân loại trong mô hình mạng nơ rơn CNN.

Chẳng hạn ta có tập các danh mục môn thể thao:



```python
sports = ['Soccer', 'Volleyball', 'Badminton', 'Basketball','Tenis', 'Golf', 'Esports']

from sklearn.preprocessing import LabelEncoder

gle = LabelEncoder()
genre_labels = gle.fit_transform(sports)
genre_mappings = {index: label for index, label in enumerate(gle.classes_)}
genre_mappings
```

Đối với dữ liệu phân loại thứ tự như các size quần áo ta chuẩn hóa như sau:

```python
sizes = ['XS', 'S', 'M', 'L', 'XL', 'XXL']

sizes_mapping = {'XS': 0, 'S': 1, 'M': 2, 'L': 3, 'XL':4, 'XXL':5}
for value, key in enumerate(sizes_mapping):
    print(key, " :  ", value)
```

Thực hiện việc xử lí trên tập data có sẵn vgsacle.cvs

```python
import pandas as pd 
import numpy as np

# read frame data

frames = pd.read_csv('vgsales.csv')
frames.iloc[:10]
```

Quan sát dữ liệu ta thấy trường Genre là dữ liệu phân loại danh mục nên ta tiến hành mã hóa chúng 

```python
from sklearn.preprocessing import LabelEncoder
gle = LabelEncoder()
genre_labels = gle.fit_transform(frames['Genre'])
genre_mappings = {index: label for index, label in 
                  enumerate(gle.classes_)}
genre_mappings
```

Thực hiện mã hóa xong tiến hành tạo 1 thêm 1 cột vào dữ liệu tên cột là 'GenreLable'

```python
frames['GenreLabel'] = genre_labels
frames[['Name', 'Platform', 'Year', 'Genre', 'GenreLabel']].iloc[1:7]
```

Đôi với dữ liệu phân loại thứ tự phương pháp mã hóa áp dụng đó sẽ là one-hot-vector (nó sẽ nắm bắt được vị trí và tính phân loại của từng danh mục). Còn dữ liệu phân loại bên trên ta không quan tâm thứ tự cái nào trước cái nào sau của dữ liệu.



```python
pokemon_frames = pd.read_csv('Pokemon.csv')
pokemon_frames.head()
```

```python
# Thực hiện phân tích trường nghi vấn Generation.
import numpy as np
unique_value = np.unique(pokemon_frames['Generation'].tolist())
unique_value
```

Nhận xét tất cả trường thông tin có 6 thế hệ từ 1 đến 6. Chúng có thứ tự và thứ hạng các thế hệ cách mã hóa one-hot-vector sẽ đc áp dụng như sau:

Trong TH các trường kia có thứ tự nhưng chưa chuẩn hóa dạng label ta cần LabelEncoder trước.




```python
le = LabelEncoder()
feature_le = le.fit_transform(pokemon_frames['Generation'].tolist())
feature_le[:10]
genre_mappings = {index: label for index, label in 
                  enumerate(le.classes_)}
genre_mappings
```

```python
# add to frame with name_col: GenLable

pokemon_frames['GenLabel'] = feature_le

pokemon_frames.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

gen_ohe = OneHotEncoder()
gen_feature_arr = gen_ohe.fit_transform(
                              pokemon_frames[['GenLabel']]).toarray()
# gen_feature_labels = list(gen_le.classes_)
gen_feature_arr
```

```python
gen_feature_labels = list(le.classes_)
gen_feature_labels 

gen_label = ['Gen' + str(i) for i in gen_feature_labels ]
gen_label
```

```python
# add onhot to frames

gen_features_one_hot = pd.DataFrame(gen_feature_arr, 
                            columns=gen_label)
gen_features_one_hot.head()
```

```python
# add vào frames_gốc ban đầu.

poke_df = pd.concat([pokemon_frames, gen_features_one_hot], axis=1)
poke_df
```

Ta thấy việc sử dụng 6 cột nhị phân mã hóa cho 6 nhãn là 0 cần thiết chúng ta có thể bỏ đi cột 1 hoặc cột 6 khi đó trường hợp rơi vào nhãn 1 hoặc 6 tau được vector 0.


```python
gen_onehot_features = pd.get_dummies(poke_df['Generation'])
gen_onehot_features
pd.concat([poke_df[['Name', 'Generation']], gen_onehot_features], 
           axis=1).iloc[4:10]
```

```python
gen_onehot_features = pd.get_dummies(poke_df['Generation'])
gen_dummy_features = gen_onehot_features.iloc[:,:-1]
pd.concat([poke_df[['Name', 'Generation']], gen_dummy_features],  
          axis=1).iloc[4:10]
```

Trong TH có quá nhiều nhãn nếu sử dụng onhot chúng ta cần n-1 vector nhị phân ta có một phuwogn pháp.


```python
from sklearn.feature_extraction import FeatureHasher
fh = FeatureHasher(n_features=6, input_type='string')
hashed_features = fh.fit_transform(frames['Genre'])
hashed_features = hashed_features.toarray()
pd.concat([frames[['Name', 'Genre']], pd.DataFrame(hashed_features)], 
          axis=1).iloc[1:7]
```
